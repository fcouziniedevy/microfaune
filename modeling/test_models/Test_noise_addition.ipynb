```python
%matplotlib notebook
from collections import Counter
import csv
from datetime import datetime
from glob import glob
import pickle
import h5py
import os

import numpy as np
from matplotlib import pyplot as plt

from sklearn.model_selection import StratifiedShuffleSplit

from skimage import util
from scipy import fftpack
import librosa
from microfaune import audio

from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.contrib.layers import group_norm
from tensorflow.math import reduce_max

from microfaune.audio import file2spec

import IPython.display as ipd
```

```python
noise_files = glob("../../../data/citeU_annotated/wav/*neg*.wav")
audio_files = glob("../../../data/warblrb10k_public_wav/wav/*.wav")
```

```python
def check_random_addition(alpha, seed=None):
    if seed is not None:
        np.random.seed(seed)
    noise_sr, noise = audio.load_wav(np.random.choice(noise_files))
    sr, data = audio.load_wav(np.random.choice(audio_files))
    if sr != noise_sr:
        if sr < noise_sr:
            noise = librosa.core.resample(noise, noise_sr, sr)
            noise_sr = sr
        else:
            data = librosa.core.resample(data, sr, noise_sr)
            sr = noise_sr

    l = min(data.size, noise.size)
    noise = noise * np.max(data) / np.max(noise)
    print(np.max(data), np.max(noise))
    new_audio = alpha*data[:l] + (1-alpha)*noise[:l]
    
    print("Noise")
    ipd.display(ipd.Audio(noise, rate=sr))
    
    print("Data")
    ipd.display(ipd.Audio(data, rate=sr))
    
    print("Combination")
    ipd.display(ipd.Audio(new_audio, rate=sr))
```

```python
check_random_addition(0.3)
```
